###  Convert Date Columns to Datetime Format

```python
df["Admission_Date"] = pd.to_datetime(df["Admission_Date"])

df["Discharge_Date"] = pd.to_datetime(df["Discharge_Date"])
```

```python
df.info()
```

### Create Length_of_Stay Feature

```python
df["Length_of_Stay"] = (
    df["Discharge_Date"] - df["Admission_Date"]
).dt.days

df.head()
```

```python
df[["Admission_Date",
    "Discharge_Date",
    "Length_of_Stay"]].head()
```

## PYTHON DATA ANALYSIS (EDA)

### Department-wise Patient Count

```python
df["Department"].value_counts()
```

```python
import matplotlib.pyplot as plt

department_count = df["Department"].value_counts()

plt.figure(figsize=(8,5))

department_count.plot(kind="bar")

plt.title("Department-wise Patient Count")
plt.xlabel("Department")
plt.ylabel("Number of Patients")

plt.show()
```

### Gender Distribution Analysis

```python
df["Gender"].value_counts()
```

```python
import matplotlib.pyplot as plt

gender_count = df["Gender"].value_counts()

plt.figure(figsize=(6,6))

plt.pie(
    gender_count,
    labels=gender_count.index,
    autopct="%1.1f%%"
)

plt.title("Gender Distribution of Patients")

plt.show()
```

### Treatment Cost Analysis by Department

```python
df.groupby("Department")["Treatment_Cost"].mean()
```

```python
avg_cost = df.groupby("Department")["Treatment_Cost"].mean()

plt.figure(figsize=(8,5))

avg_cost.plot(kind="bar")

plt.title("Average Treatment Cost by Department")
plt.xlabel("Department")
plt.ylabel("Average Treatment Cost")

plt.show()
```

### Patient Status Analysis

```python
df["Patient_Status"].value_counts()
```

```python
status_count = df["Patient_Status"].value_counts()

plt.figure(figsize=(8,5))

status_count.plot(kind="bar")

plt.title("Patient Status Distribution")
plt.xlabel("Patient Status")
plt.ylabel("Number of Patients")

plt.show()
```

### Insurance Status Analysis

```python
df["Insurance_Status"].value_counts()
```

```python
insurance_count = df["Insurance_Status"].value_counts()

plt.figure(figsize=(6,6))

plt.pie(
    insurance_count,
    labels=insurance_count.index,
    autopct="%1.1f%%"
)

plt.title("Insurance Coverage Distribution")

plt.show()
```

### Top Disease Analysis

```python
df["Disease"].value_counts()
```

```python
disease_count = df["Disease"].value_counts()

plt.figure(figsize=(10,5))

disease_count.plot(kind="bar")

plt.title("Disease-wise Patient Count")
plt.xlabel("Disease")
plt.ylabel("Number of Patients")

plt.show()
```

### Length of Stay Analysis (Healthcare KPI)

```python
df["Length_of_Stay"].describe()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df["Length_of_Stay"], bins=15)

plt.title("Length of Stay Distribution")
plt.xlabel("Length of Stay (Days)")
plt.ylabel("Number of Patients")

plt.show()
```
